### Join the 2 dataframes

```python
lego_sets1 = lego_sets.merge(parent_theme, left_on ='parent_theme',right_on='name',how='left')
```

```python
lego_sets1.shape
```

```python
lego_sets1 = lego_sets1.dropna(subset = ['set_num'])
```

### Qn 2. In which year was Star Wars not the most popular licensed theme ( in terms of number of sets released that year)?

```python
#licensed2= licensed
```

```python
licensed['num_parts'].shape
```

```python
license_sorted = licensed.sort_values('year')
license_sorted['count'] = 1
```

```python
license_sorted.head()
```

```python
sumparts = license_sorted.groupby(['year','parent_theme']).sum().reset_index()
sumparts.head(10)
```

```python
maxdf = sumparts.sort_values(by='num_parts',ascending = False).drop_duplicates(['year'])
maxdf
```

```python
maxdf.sort_values('year', inplace=True)
maxdf
```
